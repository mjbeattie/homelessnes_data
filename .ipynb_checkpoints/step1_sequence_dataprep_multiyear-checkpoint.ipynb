## Drug Use Pathways
We define a _drug use pathway_ as a sequence of drugs used by a respondent to the NSDUH survey.  The drugs in the pathway are all of the substances that the respondent indicates he has used.  The sequence is given by the age of first use (AFU) of the drugs from youngest to oldest.  We can investigate the pathways with several methods.  We will look at _pathway pairs_ to examine the most common progression from one drug to the next.  A pathway pair is given by $(i,j)$ where $i$ is the index of the originating drug and $j$ is the index of the drug immediately following drug $i$ in the pathway.  We also use the term _edge_ to describe a pathway pair.

We will also model the pathways as an adjacency matrix for a weighted graph.  This adjacency matrix is defined as follows:
$$A_{i,j}=\text{afu}_{\text{drug}_j}-\text{afu}_{\text{drug}_i}$$
$$\text{afu}_{\text{drug}_0} \equiv 0$$
$i$ and $j$ are defined the same way as in the pathway pair.

We have elected to restrict this study to adult (over 18) respondents to focus on pathways likely to have multiple nodes.

### Breaking afu ties
The afu values in the NSDUH survey are integer values.  So if a respondent started using cigarettes and alcohol at age 16, we are unable to determine which drug was the precedent.  To adjust for this, we add a random value between 0 and 1 to the afu for each drug.  This is equivalent to flipping a coin to determine which drug preceded the other.  Given the large number of responses in the survey, any bias of this method should be averaged out.

### Run the pipeline
This pipeline transforms the raw NSDUH data into a dataset that is restricted to our subjects and variables of interest.  It also generates the drug use pathways and stores them in two different ways in the dataset. The pipeline does the following:
* Removes all fields not used in this study
* Drops youth respondents
* Creates a unique identifier for each respondent.  The survey year is the set of first four characters of the identifier.
* For stimulants, tranquilizers, pain relievers, and sedatives, puts a missing value (None) into the AFU field where it is currently held by 993 and converts 999 to 991, the missing value field for all drugs.
* Consolidates cigarettes, cigars, and smokeless tobacco into one category
* Consolidates cocaine and crack cocaine into one category.
* Consolidates stimulants, tranquilizers, pain relievers, and sedatives into one category.
* Creates a small size field to store the drug use pathway
* Creates an adjacency matrix for the weighted graph that represents the pathway

We also define the years considered for the study in this section.


```python
"""
Read in NSDUH data into a pandas dataframe
"""
def dataimport(allvars, dropages, year):
    fname = datapath + 'NSDUH_' + str(year) + '_Tab.tsv'
    nsduh = pd.read_csv(fname, delimiter='\t', low_memory=False)

    # Create a pipeline to extract the variables, create a new respondent ID, and create a path of drugs in afu order.
    prep_pipe = Pipeline([('selector', pu.DataFrameSelector(allvars)),
                      ('dropages', pu.DropAges(dropages)),
                      ('dropna', pu.DataSampleDropper()),
                      ('addyear', pu.AddYearRespID(year)),
                      ('svcflag', pu.SvcFlag()),
                      ('afuset', pu.AFUSet()),
                      ('afugroup', pu.AFUGroup())])

    # Execute pipeline
    return prep_pipe.transform(nsduh)

# Set the year to readin and the agecategory to drop ([1] = under 18, [9999] = no drops)
dropages = [1]

# Execute the pipeline for the years
nsduh = []
for year in years:
    nsduh.append(dataimport(allvars, dropages, year))
```

```python
# Concatenate the data from the years into one dataframe
nsduh2 = pd.concat(nsduh, ignore_index=True, axis=0)

# Convert the observation weight into a per year count
nsduh2['YRWEIGHT'] = nsduh2['ANALWT_C']/len(years)
nsduh2 = nsduh2.drop(columns=['ANALWT_C'])

# Simple check of the data
dfname = 'nsduh2'
print('The shape of', dfname, 'is:', nsduh2.shape)
print('The index of', dfname, 'is:', nsduh2.index)
nsduh2.head()
```

### Impute missing AFU values
991 in the AFU field indicates non-usage prescription drugs.  If None is present, the respondent used the drug but NSDUH did not ask the AFU because the drug was not initiated in the past 12 months.  In that case, we impute the AFU using KNN.  

After imputation, combine the prescription drugs into one category equal to the minimum AFU among them.

```python
# Generate drug use counts
def getcounts(df, druglist):
    for item in druglist:
        df['DRUGCOUNT'] = np.where(((df[item] > 0) & (df[item] < 991)), 1, 0)
        print('Count of', item, ':', sum(df['DRUGCOUNT']))
        df.drop(['DRUGCOUNT'], axis='columns', inplace=True)
    return

druglist = ['IRPNRNMAGE', 'IRTRQNMAGE', 'IRSTMNMAGE', 'IRSEDNMAGE']
getcounts(nsduh2, druglist)
```

```python
# Impute the use without AFU pain reliever rows based upon rows with AFUs
# Copy dataframe to one with users and drop the ID column from a copy of that one
# Note:  Integer values cannot read NaN, but do read None.  We need to covert column to float to get NaN back.
def imputescrips(df, druglist):
    for item in druglist:
        scripdf = df[df[item] != 991].copy()
        scripdf_no_id = scripdf.drop(['RESPID'], axis='columns')
        scripdf_no_id = scripdf_no_id.apply(lambda x: pd.to_numeric(x, errors='coerce') )

        # Impute the missing pain reliever users.  This returns a numpy array.  Set it to a dataframe with the right column names.
        imputer = KNNImputer(n_neighbors=5)
        missnp = imputer.fit_transform(scripdf_no_id)
        missdf = pd.DataFrame(missnp, columns = scripdf_no_id.columns).astype(int)

        # Set the pain reliever column of the initial copy equal to that of the imputed dataframe
        scripdf[item] = missdf[item].values

        # Make a copy of the original dataframe without the pain reliever AFU and concatenate with the final imputed df
        df2 = df[df[item] == 991]
        df = pd.concat([df2, scripdf], axis=0)
        print('Missing values:\n', df.isna().any()[lambda x: x])
    return df
```

```python
nsduh3 = imputescrips(nsduh2, druglist)

# Recount the positive uses for drugs
getcounts(nsduh3, druglist)
```

```python
# Set the IRSCRIPAGE variable to the minimum of the given prescription drug AFU values and drop the original prescription columns
nsduh3['IRSCRIPAGE'] = nsduh3.loc[:, ['IRPNRNMAGE', 'IRTRQNMAGE', 'IRSTMNMAGE', 'IRSEDNMAGE']].min(axis=1)
nsduh3 = nsduh3.drop(columns=['IRPNRNMAGE', 'IRTRQNMAGE', 'IRSTMNMAGE', 'IRSEDNMAGE'])
getcounts(nsduh3, ['IRSCRIPAGE'])
```

### Print out the most common edges and paths found in the NSDUH dataset
We will print out the edges and paths that make up 50% of the total unique edges and 40% of the total unique paths.  Note that there are many more unique paths than unique edges.

```python
# Generate a dataframe containing all of the different edges and their frequency
# This block takes the edges in the paths for each row and stores them and their row weight
afuGraphDF = nsduh3[['RESPID','YRWEIGHT','AFUPATH']]
edgelist = []
for row in afuGraphDF.itertuples():
    path = row.AFUPATH
    if len(path)==1:
#        edgelist.append([(path[0][0],'NONE'),row.YRWEIGHT])
        edgelist.append([('start','NONE'),row.YRWEIGHT])
    else:
        for i in range(0,len(path)-1):
            edgelist.append([(path[i][0],path[i+1][0]),row.YRWEIGHT])

# This block summarizes the counts of the edges
edgelistdf = pd.DataFrame (edgelist, columns = ['edge','weight'])
edgelistdf = edgelistdf.groupby('edge').sum('weight').sort_values(by='weight', ascending=False)
totaledges = edgelistdf['weight'].sum()
edgelistdf['fraction'] = edgelistdf['weight']/totaledges
edgelistdf.reset_index(inplace=True)
```

```python
print('The number of total edges is', totaledges)
print('The number of unique edges is', edgelistdf.shape[0])
edgelistdf.head()
```

```python
# Print out edges to output
filename = outpath + 'edges_and_paths.txt'
fileout = open(filename, 'w')
print('EDGELIST ENUMERATION', file=fileout)
print('The total number of edges is:', totaledges, file=fileout)
print('The total number of unique edges is:', edgelistdf.shape[0], file=fileout)
totfrac = 0
for row in edgelistdf.itertuples():
    totfrac += row.fraction
    print(row.edge, '\t', round(row.weight,0), '\t', round(row.fraction,6), '\t', round(totfrac,6), file=fileout)
    
```

```python
# Generate a dataframe containing all of the different paths and their frequency
# This block takes the the paths for each row and stores them and their row weight
afuGraphDF = nsduh3[['RESPID','YRWEIGHT','UNWAFUPATH']]
pathlist = []
for row in afuGraphDF.itertuples():
    path = str(row.UNWAFUPATH)
    pathlist.append([path,row.YRWEIGHT])

# This block summarizes the counts of the edges
pathlistdf = pd.DataFrame (pathlist, columns = ['path','weight'])
pathlistdf = pathlistdf.groupby('path').sum('weight').sort_values(by='weight', ascending=False)
totalpaths = pathlistdf['weight'].sum()
pathlistdf['fraction'] = pathlistdf['weight']/totalpaths
pathlistdf.reset_index(inplace=True)
```

```python
print('The number of total paths is', totalpaths)
print('The number of unique paths is', pathlistdf.shape[0])
pathlistdf.head()
```

```python
# Print the most common paths
print('\n\n\nMOST COMMON PATHS', file=fileout)
print('The total number of paths is:', totalpaths, file=fileout)
print('The total number of unique is:', pathlistdf.shape[0], file=fileout)
totfrac = 0
for row in pathlistdf.itertuples():
    totfrac += row.fraction
    if totfrac <= 0.8:
        print(row.path, '\t', round(row.weight,0), '\t', round(row.fraction,6), '\t', round(totfrac,6), file=fileout)

fileout.close()
```

## Visualization of edges
We create a weighted directed graph of the unique edges.  Each node of the graph is a drug, and the edge between two nodes represents the sequence of afu of the two drugs.  The tail of the arrow is at the drug used at the lower age, and the head of the arrow is at the drug used at the higher age.  The weights of the edges correspond to the log number of times the edge appears in the respondents' pathways.  In this visualization, the darkness of the edge increases with its weight.

```python
# Add columns to fit NetworkX format:  node1, node2, attributes.  Take log fraction to improve graphing
def getnode(row, index):
    return row.edge[index]

edgelistdf['node1'] = edgelistdf.apply(lambda row: getnode(row, 0), axis=1)
edgelistdf['node2'] = edgelistdf.apply(lambda row: getnode(row, 1), axis=1)
edgelistdf['logfrac'] = np.log(edgelistdf['fraction'])
edgelistdf.head()
```

```python
# Read the pandas edgelist into NetworkX
G = nx.from_pandas_edgelist(edgelistdf, source='node1', target='node2', edge_attr=['fraction','logfrac'],create_using=nx.DiGraph())

# Draw the graph with darker colors for heavier weights
bboxdict = dict(facecolor='blue')
edges,weights = zip(*nx.get_edge_attributes(G,'logfrac').items())

figname = outpath + 'drugpair_graph.jpg'
fig, ax1 = plt.subplots(1, 1, figsize=(20,10))
fig.suptitle('Drug Use Pathway Pairs (Lower Age to Higher Age)\nEdges Weighted by log(Fraction)', fontsize=22)
pos = nx.circular_layout(G)
nx.draw_networkx_labels(G, pos, font_color='brown', font_size=24, font_family="sans-serif", ax=ax1)
nx.draw(G, pos, node_color='white', edgelist=edges, edge_color=weights, width=3.5, edge_cmap=plt.cm.binary, 
        connectionstyle='arc3, rad = 0.1', ax=ax1)
plt.savefig(figname)
plt.show()

```
